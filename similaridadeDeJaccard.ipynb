```
import networkx as nx
```

```
#questão 3
from networkx.algorithms.link_prediction import jaccard_coefficient
G = nx.Graph()
list = ['a','b','c','d','e','f','g','h','i','j']
G.add_nodes_from(list)
aresta= [ ('a','b'),('a','c'),('a','h'),('b','c'),('b','h'),('c','d'),('c','h'),('d','e'),('e','f'),('f','g'),('f','i'),('f','j'),('g','i'),('g','j'),('i','j')]
G.add_edges_from(aresta)
resp =nx.jaccard_coefficient(G,[('a','b')])
resp2 =nx.jaccard_coefficient(G,[('b','c')])
resp3 =nx.jaccard_coefficient(G,[('c','d')])
resp4 =nx.jaccard_coefficient(G,[('d','e')])
nx.draw(G ,with_labels =True)
for u,v,p in resp:
     print(f"({u}, {v}) -> {p:.4f}")
for u,v,p in resp2:
       print(f"({u}, {v}) -> {p:.4f}")  
for u,v,p in resp3:
       print(f"({u}, {v}) -> {p:.4f}") 
for u,v,p in resp4:
       print(f"({u}, {v}) -> {p:.4f}") 
```
