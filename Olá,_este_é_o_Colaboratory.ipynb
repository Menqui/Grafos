```
import networkx as nx
```

```
 G1 = nx.Graph()
 G2 = nx.Graph()

 # adicianando vertices e arestas
 list_a = ['A','B','C','D','E','F','G','H']
 G1.add_nodes_from(list_a)
 arestas_a = [('A','H'), ('A','B') , ('A','F') , ('A','E') , ('A','D'), ('B','C'), ('B','G'), ('B','F') , ('B','E'), ('C','D') , ('C','H') , ('C','H') , ('C','F') , ('D','E'), ('D','H') , ('D','G') , ('E','F') , ('E','H') , ('F','G')]
 G1.add_edges_from(arestas_a)
 nx.draw(G1,with_labels = True , node_color = 'purple') 
```

```
list_b = [1,2,3,4,5,6,7,8]
G2.add_nodes_from(list_b)
aresta_b = [(1,8),(1,2),(1,7),(1,3),(1,5),(8,7),(8,6),(8,2),(8,4),(7,3),(7,5),(2,6),(6,4),(6,5),(6,7),(5,4),(5,3),(4,2),(3,4),(3,2)]
G2.add_edges_from(aresta_b)
nx.draw(G2,with_labels=True , node_color = 'red')
```

```
nx.is_isomorphic(G1,G2)
```

```
# subgrafos
 #questão 2

q3 = nx.Graph()
q4 = nx.Graph()
array_lista = ['000 ',' 001' , '010', '100', '011' , ' 111']
q3.add_nodes_from(array_lista)
array_listb = ['0000','0001','0010','0100','1000','0011','0111','1100','1010','1110','1111','1011']
q4.add_nodes_from(array_listb)
#nx.draw(q3,with_labels =True , node_color = 'red')
#nx.draw(q4,with_labels=True)

subgrafos = [q3.subgraph(q4) for q4 in nx.connected_components(q3)]
print(len(subgrafos)) #printa quantidade de subgrafos
```

```
nx.draw(
    subgrafos[len(subgrafos)-1],       # subgrafo na posição final(q4)
    with_labels=True,   # mostrar rótulos
    node_size=800,      # aumentar tamanho
    node_color='red'    # cor vermelha
)
#mostra os subgrafos de q3 em q4
```
