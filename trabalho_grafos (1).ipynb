```
#cria o grafo
grafo_a=nx.Graph()
x=int(x)#converte o valor de x para inteiro
#recebe os vertices
for j in nu:
  if j%1 ==0: #se o num difere de um num quebrado
   vertices.append(j)
print(vertices)
#cria os vertices e as arestas
grafo_a.add_nodes_from(vertices)
#arestas
#separar as tuplas das listas de duas em duas
n=2
arestas=[tuple(vertices[i:i + n]) for i in range(0, len(vertices), n)]
grafo_a.add_edges_from(arestas) #adiciona as arestas
print(arestas)
nx.draw(grafo_a,with_labels=True)
```
